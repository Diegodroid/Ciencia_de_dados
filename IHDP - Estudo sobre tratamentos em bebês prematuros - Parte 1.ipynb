# 1.   Introdução






A prematuridade, caracterizada pelo nascimento de um bebê antes de completar 37 semanas de gestação, é um dos desafios mais significativos enfrentados pela medicina neonatal e pela saúde pública em todo o mundo. O aumento da taxa de nascimentos prematuros nas últimas décadas tem suscitado crescente preocupação, uma vez que esses bebês podem enfrentar uma série de problemas médicos e desenvolvimentais, bem como desafios emocionais para suas famílias.

Os bebês prematuros são especialmente vulneráveis devido à imaturidade de seus órgãos e sistemas, que normalmente continuariam a se desenvolver dentro do útero da mãe. Essa imaturidade pode levar a complicações de curto e longo prazo, afetando não apenas o início da vida, mas também repercutindo na infância, adolescência e até mesmo na vida adulta.

Dentre os problemas imediatos associados à prematuridade, destacam-se a dificuldade respiratória, a susceptibilidade a infecções, a instabilidade térmica e a alimentação desafiadora. Além disso, bebês prematuros têm maior probabilidade de desenvolver complicações neurológicas, como hemorragias intracranianas e distúrbios do desenvolvimento cerebral, que podem afetar seu crescimento e aprendizado.

Alguns estudos e programas foram desenvolvidos para combater os efeitos da prematuridade, entre eles, podemos citar o  Programa de Saúde e Desenvolvimento do Bebê (IHDP), que foi um estudo sobre intervenção em bebês com baixo peso ao nascer (BPN) realizado na década de 1980 - representa um marco significativo. Trata-se de uma iniciativa pioneira, sendo o primeiro experimento clínico randomizado conduzido em diversos locais, com o objetivo de avaliar a eficácia da combinação de serviços destinados ao desenvolvimento infantil e apoio familiar, juntamente com acompanhamento pediátrico, visando reduzir problemas de desenvolvimento, comportamentais e de saúde em prematuros BPN. (Ross, 2008).

Este trabalho tem como objetivo explorar de forma mais aprofundada os problemas associados à prematuridade em bebês, analisando suas possíveis causas.

# 2.   Referencial Teórico






De acordo com estimativas, ocorre anualmente em todo o mundo o nascimento de 15 milhões de bebês prematuros (Who, 2018). O Brasil, por sua vez, ocupa o 9º lugar em números absolutos de nascimentos prematuros, tornando essa questão um relevante problema de saúde pública (Blencowe et.al., 2012) . Conforme a definição da Organização Mundial da Saúde (OMS), são considerados neonatos prematuros aqueles que nascem antes de completar as 37 semanas de gestação. Com base na idade gestacional, esses recém-nascidos podem ser classificados ainda em subcategorias, tais como pré-termos extremos (até 28 semanas incompletas), muito prematuros (de 28 semanas completas a 32 semanas incompletas) e prematuros moderados a tardios (de 32 semanas incompletas a 37 semanas incompletas)(Who, 2018). Essa classificação é de extrema importância para avaliar o grau de prematuridade e auxiliar no planejamento de cuidados adequados a cada grupo de bebês prematuros, visto que essas crianças têm necessidades específicas de acordo com o estágio de desenvolvimento em que se encontram ao nascer. Diante dessa significativa incidência de prematuridade, torna-se imperativo que medidas de prevenção e políticas de saúde sejam adotadas com o objetivo de reduzir os riscos associados a essa condição e garantir o melhor prognóstico possível para esses recém-nascidos, promovendo assim a saúde materno-infantil e o bem-estar global da população.

Ramos e Cuman (2009), buscaram identificar o perfil das mães e dos bebês prematuros nascidos vivos, assim como caracterizar os recém-nascidos prematuros em situação de risco para o crescimento e desenvolvimento. A pesquisa epidemiológica de corte transversal foi conduzida na cidade de Guarapuava, no estado do Paraná. Os dados foram obtidos por meio do sistema de informações sobre nascidos vivos (SINASC), com base nas declarações de nascidos vivos. A análise estatística foi realizada em uma amostra composta por 106 declarações de nascidos vivos prematuros, no período de janeiro a junho de 2005, juntamente com informações sobre suas respectivas mães.

Foram consideradas as seguintes variáveis para análise: características sociodemográficas, condições da gestação e do parto, bem como as características dos prematuros nascidos vivos. Os resultados obtidos permitiram concluir que conhecer e avaliar o perfil das mães, assim como o número e a situação dos nascimentos de bebês prematuros em uma determinada área e período de tempo, é de suma importância para a identificação de riscos vitais relacionados às condições do nascimento, crescimento e desenvolvimento infantil.

Os autores concluíram que o perfil das mães dos bebês prematuros e a caracterização dos nascidos vivos são diretamente influenciados pelas condições socioeconômicas e sanitárias do local onde ocorre a gestação e o parto. Essas condições, por sua vez, têm um impacto significativo na qualidade de vida futura dessas crianças, tornando-se um fator contribuinte para os altos índices de mortalidade infantil, que é um dos principais problemas observados no município estudado.

No que condiz aos fatores de riscos, podemos destacar os seguintes estudos. FERREIRA JUNIOR et al., (2018) mostrou que foi observada a predominância do sexo masculino (53,3%) entre os nascimentos prematuros.

O estudo conduzido por Kawakita, Wilson, Grantz, Landy, Huang e Gomez-Lobo (2016) investigou uma amostra de 43.537 mulheres com idade inferior a 25 anos nos Estados Unidos, com o objetivo de analisar a relação entre a idade materna e o risco de parto prematuro.

Os resultados da pesquisa demonstraram que existe uma associação significativa entre a idade materna mais jovem e um aumento nas chances de ocorrer parto prematuro. Em outras palavras, quanto mais jovem a mulher, maior é a probabilidade de ela dar à luz um bebê prematuro.

A idade materna é um fator reconhecido como condicionante de risco durante a gravidez, especialmente quando ocorre fora da faixa etária considerada ideal para a gestação, ou seja, nos extremos da vida reprodutiva, abrangendo tanto as gestações precoces quanto as tardias (Cabral; Santos & Cano, 2017).

Em mulheres com gestação tardia, tem sido observado um aumento no número de abortos espontâneos e induzidos, bem como um maior risco para a mortalidade perinatal, baixo peso ao nascer, parto pré-termo e recém-nascidos com baixo peso em relação à idade gestacional. As gestações em mulheres de idade materna avançada também são consideradas de alto risco, devido, principalmente, à crescente incidência de síndromes hipertensivas, ruptura prematura de membranas e presença de diabetes gestacional. Além disso, existe uma maior chance de o Índice de Apgar no quinto minuto ser inferior a sete, o que pode indicar a ocorrência de anomalias cromossômicas, macrossomias (bebês com peso acima do esperado para a idade gestacional), baixo peso ao nascer e menor vitalidade do neonato. Sendo assim, um escore menor que 7 no Índice de Apgar é considerado um sinal de alerta para a necessidade de uma atenção especial e cuidados adicionais ao recém-nascido (Silva; Cavalcante; Carneiro & Oliveira, 2020).

É fundamental que o uso de tabaco, bebidas alcoólicas e drogas ilícitas seja suspenso durante a gestação, porém, muitas vezes, isso não ocorre. Um estudo realizado com 61 gestantes revelou que a maioria das mulheres que eram fumantes e ex-fumantes não tinha conhecimento dos efeitos prejudiciais que o tabagismo pode causar ao feto e ao recém-nascido, nem das doenças relacionadas a essa prática (Bertani; Garcia; Tanni & Godoy, 2015).

Diante dessa realidade, torna-se essencial que os profissionais da área da saúde ofereçam orientação adequada sobre os possíveis riscos associados ao tabagismo, assim como os perigos da exposição à radiação durante a gestação. Além disso, é importante alertar sobre os malefícios do consumo de bebidas alcoólicas durante a gravidez, pois o álcool pode levar a abortamento, restrição do crescimento fetal, retardo mental e malformações no bebê.

O retardo mental é considerado uma das principais complicações que o álcool pode provocar no recém-nascido, e é fundamental destacar que essa condição pode ser totalmente prevenida (Sociedade de Pediatria de São Paulo, 2017).

Promover a conscientização sobre essas questões é fundamental para proteger a saúde da mãe e do bebê durante a gestação, garantindo uma gravidez mais saudável e segura. O papel dos profissionais de saúde é crucial nesse sentido, pois sua orientação e suporte podem ajudar as gestantes a tomarem decisões mais saudáveis para si mesmas e para seus filhos.

No que diz respeito à renda familiar mensal, o estudo realizado por Almeida, et al (2012), com 58 puérperas de bebês prematuros evidenciou que 89% delas tinham uma renda mensal de menos de dois salários-mínimos. Essa variável, isoladamente ou em conjunto com outros fatores, foi identificada como um fator que aumenta o risco de partos prematuros.

Conforme amplamente documentado na literatura, a baixa escolaridade materna está associada a nascimentos prematuros e pode ser influenciada pelo menor acesso à informação e aos serviços de saúde. O estudo de Oliveira, et al.(2016) destaca que mulheres com escolaridade inadequada para a idade têm duas vezes mais chances de terem partos prematuros. Além disso, Bernardino, et al. (2019) também observou uma prevalência de mães de bebês prematuros com apenas o ensino médio completo ou ainda incompleto.

O acompanhamento pré-natal é fundamental para identificar e investigar situações de risco durante a gestação, prevenindo desfechos indesejáveis. No estudo de Almeida et al. (2012), com 58 mães de bebês prematuros, cerca de 60% realizaram menos de cinco consultas pré-natal, apresentando cinco vezes mais chances de prematuridade em comparação com as mães de bebês a termo. Outros estudos, como os de Oliveira et al. (2016), também relacionaram o número inadequado de consultas pré-natal com o parto prematuro.

Esses resultados enfatizam a importância da atenção à saúde materna durante a gravidez, especialmente para as mulheres que apresentam baixa renda, baixa escolaridade e acesso limitado aos cuidados pré-natais. O fortalecimento do acompanhamento pré-natal adequado pode contribuir para reduzir os riscos de partos prematuros e melhorar os resultados de saúde tanto para a mãe quanto para o bebê. Ações que visem melhorar o acesso a informações e serviços de saúde, bem como programas de conscientização sobre a importância do pré-natal, são essenciais para abordar esses fatores de risco e promover uma gravidez mais segura e saudável.

## Análise de Possíveis *Outliers*

Essa Análise se dará separando os tipos de tratamentos. Neste trabalho será considerado Outliers valores 1,5 * FIQ (Faixa Interquartil)  a baixo do primeiro quartil, ou acima do terceiro quartil.  

```r
# Exibe as informações estatísticas básica das variáveis
print("GRUPO DE TRATAMENTO Caso")
summary(CASO)
print("GRUPO DE TRATAMENTO Controle")
summary(CTRL)


```

```r
#IDADE

print("GRUPO CASO")
## CASO
Q1 <- quantile(CASO$IDADE, probs = 0.25) #primeiro quartil
Q3 <- quantile(CASO$IDADE, probs = 0.75) # terceiro quartil
IQR <- Q3 - Q1 #indice
LimInf <- Q1 - 1.5*IQR #limite inferior
LimSup <- Q3 + 1.5*IQR #limite superior
OutlierIdade <- subset(CASO, IDADE < LimInf | IDADE > LimSup)
OutlierIdade

print("GRUPO CONTROLE")
## CONTROLE
Q1 <- quantile(CTRL$IDADE, probs = 0.25)
Q3 <- quantile(CTRL$IDADE, probs = 0.75)
IQR <- Q3 - Q1
LimInf <- Q1 - 1.5*IQR
LimSup <- Q3 + 1.5*IQR
OutlierIdade <- subset(CTRL, IDADE < LimInf | IDADE > LimSup)
OutlierIdade
```

Foi encontrado apenas um dado, para o grupo CASO, com um valor extremo por idade, contudo, essa observação vai de encontro com uma das hipóteses abordadas no trabalho, que mulheres com idades acima de 40 anos, tem uma maior possibilidade de ter um filho prematuro. Vale ressaltar que essa mãe nao fez uso de cigarro, alcool, e realizou o pré-natal.

```r
#PESO
print("GRUPO CASO")
## CASO
Q1 <- quantile(CASO$PESO, probs = 0.25)
Q3 <- quantile(CASO$PESO, probs = 0.75)
IQR <- Q3 - Q1
LimInf <- Q1 - 1.5*IQR
LimSup <- Q3 + 1.5*IQR
Outlier <- subset(CASO, PESO < LimInf | PESO > LimSup)
Outlier
print("GRUPO CONTROLE")
## CONTROLE
Q1 <- quantile(CTRL$PESO, probs = 0.25)
Q3 <- quantile(CTRL$PESO, probs = 0.75)
IQR <- Q3 - Q1
LimInfCTR <- Q1 - 1.5*IQR
LimSupCTR <- Q3 + 1.5*IQR
Outlier <- subset(CTRL, PESO < LimInfCTR | PESO > LimSupCTR)
cat("numero de observacoes de outliers para o Peso do Grupo Controle",
nrow(Outlier))
```

Na variável Peso, não foi obtido nenhum outlier para o grupo de tratamento do CASO, contudo no CONTROLE tiveram 78 observações. Abaixo, será analisado o motivo das 78 observações.

```r
#PESO

#abaixo de 2500 que e o limite para entrar no tratamento caso
print("Grupo Controle")
PesoBaixo <- subset(CTRL, PESO < 2500 & (PREMATCAT != "postermo" &
  PREMATCAT != "ausente"))
#abaixo ao limite inferior de outlier do grupo controle
PesoBaixoOutLier <- subset(CTRL, PESO < LimInfCTR & (PREMATCAT != "postermo" &
  PREMATCAT != "ausente"))
#abaixo ao limite inferior de outlier do grupo controle, considerando sem
#nenhuma prematuridade
PesoBaixoOutLierNoPremat <- subset(CTRL, PESO < LimInfCTR & (PREMATCAT
  == "postermo" | PREMATCAT == "ausente"))
cat("n° de observações para peso < 2500 e foram prematuros:" ,nrow(PesoBaixo))
cat('\n')
cat("n° de observações de outliers para peso < 2500 que foram prematuros:" ,
    nrow(PesoBaixoOutLier))
cat('\n')
print("Análise Estatística  de Peso para os Outliers de prematuridade postermo
  ou ausente para peso menor que 2500kg")
summary(PesoBaixoOutLierNoPremat)
```

Essas conclusões destacam a importância de considerar a etnia, a idade e a renda das mulheres no contexto da prematuridade, bem como os possíveis efeitos do uso de cigarro e álcool durante a gravidez e o acompanhamento pré-natal. No entanto, é essencial lembrar que essas conclusões são baseadas em análises preliminares e, para obter conclusões mais robustas e precisas, é fundamental realizar estudos estatísticos mais aprofundados, controlando outros possíveis fatores de confusão e ampliando a amostra para uma análise mais abrangente.
